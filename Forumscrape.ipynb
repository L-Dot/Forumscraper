```python
from selenium import webdriver
import time

driver = webdriver.Firefox()

driver.get('https://forum.nl/nl/agenda?categorie=films')

time.sleep(1)
previous_height = driver.execute_script('return document.body.scrollHeight')

while True:
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    
    time.sleep(1)
    
    new_height = driver.execute_script('return document.body.scrollHeight')
    if new_height == previous_height:
        break
    previous_height = new_height
```

```python
films = []

# extracting a list of all days
day_elements = driver.find_elements_by_xpath("//div[contains(@class,'col-1-1 no-p calendar-day overview paging-container active')]")

for day in day_elements[:10]:
    
    # get the date
    date = day.get_attribute("data-last-date")        
    
    # get a list of all film elements of that day
    film_elements = day.find_elements_by_xpath(".//a[contains(@class,'ticket-row overview space-b')]")
    
    film_prop = {}
    for film in film_elements:
        title = film.find_element_by_xpath(".//span[contains(@class, 'main')]").text
        genre = film.find_element_by_xpath(".//span[contains(@class, 'sub')]").text
        tijd = film.find_element_by_xpath(".//div[contains(@class, 'time')]").text
        prijs = film.find_element_by_xpath(".//div[contains(@class, 'ticket')]").text
        plaatje = film.find_element_by_xpath(".//div[contains(@class, 'image bg-image')]").get_attribute('style')[23:-3]
        
        film_prop['Datum'] = date
        film_prop['Titel'] = title
        film_prop['Genre'] = genre
        film_prop['Tijd'] = tijd
        film_prop['Prijs'] = prijs
        film_prop['Plaatje'] = plaatje
        
        films.append(film_prop)
```
